```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the image
image = cv2.imread('sample.jpg')  # Replace with your image file

# Convert BGR to RGB for displaying with matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the image
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis('off')
plt.show()
```

```python
# Write the image to disk
cv2.imwrite('output_saved_image.jpg', image)
print("Image saved as 'output_saved_image.jpg'")
```

```python
# Define the cropping coordinates [y1:y2, x1:x2]
cropped_image = image[100:300, 200:400]

# Convert cropped image to RGB for display
cropped_rgb = cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB)

# Display cropped image
plt.imshow(cropped_rgb)
plt.title("Cropped Image")
plt.axis('off')
plt.show()
```

```python
# Convert to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Display grayscale image
plt.imshow(gray_image, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')
plt.show()
```

```python
# Define Sobel edge detection kernel (horizontal edges)
sobel_x = np.array([[-1, 0, 1],
                    [-2, 0, 2],
                    [-1, 0, 1]])

# Apply convolution using cv2.filter2D
edges_x = cv2.filter2D(gray_image, -1, sobel_x)

# Display edge-detected image
plt.imshow(edges_x, cmap='gray')
plt.title("Edge Detection (Sobel - X)")
plt.axis('off')
plt.show()
```

```python
# Canny edge detection
canny_edges = cv2.Canny(gray_image, threshold1=100, threshold2=200)

# Display Canny edges
plt.imshow(canny_edges, cmap='gray')
plt.title("Edge Detection (Canny)")
plt.axis('off')
plt.show()
```
